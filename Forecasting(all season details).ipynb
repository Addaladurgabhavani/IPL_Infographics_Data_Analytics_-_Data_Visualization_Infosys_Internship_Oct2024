```python
details_df = df[['season', 'match_id', 'batsman1_name', 'runs', 'isBoundary', 'isWide', 'isNoball']]
details_df
```

```python
details_df= details_df.dropna(subset=['season', 'batsman1_name', 'runs'])

# Fix data types for consistency
details_df['season'] = details_df['season'].astype(int)
details_df['isBoundary'] = details_df['isBoundary'].fillna(0).astype(bool)
details_df['isWide'] = details_df['isWide'].fillna(0).astype(bool)
```

```python
grouped = details_df.groupby(['season', 'batsman1_name']).agg(
    total_runs=('runs', 'sum'),
    total_matches=('match_id', 'nunique'),
    balls_played=('isWide', lambda x: (~x).sum()),  # Count non-wide balls
    boundaries=('isBoundary', 'sum')  # Count boundary balls
).reset_index()
```
