```python
3.TOP 15 PARTNERSHIPS BY TOTAL RUNS
```

```python
partnerships = data.groupby(['batsman1_name', 'batsman2_name'])['runs'].sum().reset_index()
#Sort and get the top 10 partnerships by total runs
top_15_partnerships = partnerships.nlargest(15, 'runs')
# Combine batsman names for display
top_15_partnerships['partnership'] = top_15_partnerships['batsman1_name'] + " & " + top_15_partnerships['batsman2_name']
plt.figure(figsize=(12, 6))
plt.barh(top_15_partnerships['partnership'], top_15_partnerships['runs'], color='green')
plt.xlabel("Total Runs")
plt.title("Top 15 Partnerships by Total Runs")
plt.gca().invert_yaxis()
plt.show()
```

```python
4.BATSMAN VS BOWLER ANALYSIS
```

```python
batsman_bowler_runs = data.groupby(['batsman1_name', 'bowler1_name'])['runs'].sum().reset_index()
# Get the top 15 batsman-bowler pairs by total runs scored
top_15_batsman_bowler_pairs = batsman_bowler_runs.nlargest(15, 'runs')
# Combine batsman and bowler names for display
top_15_batsman_bowler_pairs['pair'] = top_15_batsman_bowler_pairs['batsman1_name'] + " vs " + top_15_batsman_bowler_pairs['bowler1_name']
plt.figure(figsize=(12, 6))
plt.barh(top_15_batsman_bowler_pairs['pair'], top_15_batsman_bowler_pairs['runs'], color='green')
plt.xlabel("Total Runs Scored")
plt.title("Top 15 Batsman vs Bowler Pairs by Runs Scored")
plt.gca().invert_yaxis()
plt.show()
```

```python
5.WICKETS BY DISMISSAL KIND PER BOWLER
```

```python
wickets_data = data[data['wicket_id'].notna()]
# Group by bowler and dismissal kind, then count the number of occurrences
dismissal_counts = wickets_data.groupby(['bowler1_name', 'wkt_text']).size().reset_index(name='wicket_count')
# Select the top 10 combinations based on wicket count
top_dismissals = dismissal_counts.nlargest(15, 'wicket_count')
plt.figure(figsize=(12, 6))
plt.barh(top_dismissals.apply(lambda x: f"{x['bowler1_name']} ({x['wkt_text']})", axis=1), top_dismissals['wicket_count'], color='green')
plt.xlabel("Number of Wickets")
plt.ylabel("Bowler and Dismissal Kind")
plt.title("Top 15 Bowler and Dismissal Kind Combinations")
plt.show()
```
