# 1.AVERAGE RUN RATE PER OVER

```python
avg_runrate = df.groupby('over')['total_runs'].mean()
print(avg_runrate)

plt.figure(figsize=(12,6))
avg_runrate.plot( kind = 'line', color = 'blue', marker = 'o')
plt.title("Average Run Rate")
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.show()
```

# 2.WICKET FALL DISTRIBUTION BY OVER

```python
wickets_df = df[df['is_wicket'] == 0]
wicket_distribution = wickets_df.groupby('over').size().reset_index(name = 'wickets')
print(wicket_distribution)

plt.figure(figsize=(10,10))
plt.bar(wicket_distribution['over'], wicket_distribution['wickets'], color='blue')
plt.xlabel('Over')
plt.ylabel('Number of Wickets')
plt.title('Wicket Fall Distribution by Over')
plt.grid(axis='x', linestyle='-', alpha=0.2)
plt.show()
```

# 3.Partnership Analysis (Top 15 Partnerships by Total Runs)

```python
partnerships = df.groupby(['batter', 'non_striker'])['total_runs'].sum().reset_index()
top_15_partnerships = partnerships.nlargest(15, 'total_runs')
top_15_partnerships['partnership'] = top_15_partnerships['batter'] + ' & ' + top_15_partnerships['non_striker']
print(top_15_partnerships)
plt.figure(figsize=(12, 6))
plt.barh(top_15_partnerships['partnership'], top_15_partnerships['total_runs'], color='blue')
plt.title('Top 15 Batting Partnerships by Runs Scored')
plt.ylabel('Batting Partnership')
plt.xlabel('Total Runs')
plt.gca().invert_yaxis()
plt.show()
```

# 4.BATSMAN VS BOWLER ANALYSIS

```python
batsman_bowler = df.groupby(['batter', 'bowler'])['batsman_runs'].sum().reset_index()
top_15_bat_bowl = batsman_bowler.nlargest(15, 'batsman_runs')
top_15_bat_bowl['batsman_bowler'] = top_15_bat_bowl['batter'] + ' vs ' + top_15_bat_bowl['bowler']
print(top_15_bat_bowl)


plt.figure(figsize=(12, 6))
plt.barh(top_15_bat_bowl['batsman_bowler'], top_15_bat_bowl['batsman_runs'], color='blue')
plt.title('Top 15 Batsman vs Bowler Matchups by Runs Scored')
plt.xlabel('Total Runs')
plt.ylabel('Batsman vs Bowler')
plt.gca().invert_yaxis()
plt.show()
```
